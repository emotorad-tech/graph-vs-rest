```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Local test
#### Test size: 50,000

```python
df_graphql_local = pd.read_csv('./graphql-test/local_data.csv')
df_rest_local = pd.read_csv('./rest-test/local_data.csv')
```

```python
y = df_graphql_local['Values']
x = range(len(y))
plt.plot(x, y)
plt.xlabel('readings')
plt.ylabel('time (ns)')
plt.title('GraphQL test')
```

```python
y = df_rest_local['Values']
x = range(len(y))
plt.plot(x, y)
plt.xlabel('readings')
plt.ylabel('time (ns)')
plt.title('Rest test')
```

```python
y1 = df_graphql_local["Values"]
y2 = df_rest_local["Values"]
x = range(len(y1))
plt.plot(x, y1, label='GraphQL')
plt.xlabel('Readings')
plt.ylabel('Time (ns)')
plt.plot(x, y2, label='Rest')
plt.legend()
```

```python
print(df_graphql_local.mean())
print(df_rest_local.mean())
```
